```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('csvfile.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop(['Trust Rating'], axis=1)
y = df['Trust Rating']
```

```python
X.head()
```

```python
y.head()
```

```python
y_test.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10, shuffle=True)
```

### Linear Regression

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
y_prediction = model.predict(X_train)
```

```python
print("MAE on train data= " , metrics.mean_absolute_error(y_train, y_prediction))
```

```python
y_prediction = model.predict(X_test)
```

```python
print("MAE on test data = " , metrics.mean_absolute_error(y_test, y_prediction))
```

```python
plt.figure(figsize = (10,5))
plt.scatter(y_test, y_prediction)
plt.xticks(np.arange(0, 101, 5.0))
plt.xlabel('Actual Trust Rating')
plt.ylabel('Predicted Trust Rating')
plt.title('Actual vs Predicted Trust Rating')
plt.show()
```

```python
residuals = y_test - y_prediction
plt.scatter(y_prediction, residuals)
plt.xlabel('Predicted Trust Rating')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.scatter(range(95), y_test, color = 'red')
plt.scatter(range(95), y_prediction, color = 'blue')
plt.xticks(np.arange(0, 95, 2.0))
plt.title('Test vs Predicted Trust Rating')
plt.xlabel('Sample index')
plt.ylabel('Trust Rating')
plt.show()
```
